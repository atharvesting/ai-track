Functions are first class citizens which means that they can be treated just like any other type of variable.

```python
# Assigning function to a variable
def greet(name):
    return f"Hello, {name}!"

say_hi = greet  # leading brackets are not used

print(greet("Alice"))
print(say_hi("Hames"))
```

```python
# Passing a function as an argument
def apply_operation(func, x, y):
    return func(x, y)

def add(a, b):
    return a + b

def multiply(a, b):
    return a * b

result_add = apply_operation(add, 5, 3)
print(f"Applying add: {result_add}")
```
