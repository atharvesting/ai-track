- Set the value of the index column to the Date column in the csv file
- Telling the notebook to understand each row index as a date.

```python
import pandas as pd

file_path = 'data/fifa.csv'
df = pd.read_csv(file_path, index_col="Date", parse_dates=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,6))
sns.lineplot(data=df)
```
