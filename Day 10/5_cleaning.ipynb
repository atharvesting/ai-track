```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/NFL Play by Play 2009-2016 (v3).csv')
```

In order to find the count of rows with missing values,
- DataFrame.isnull() = returns rows that contains missing values
- DataFrame.sum() = returns the number of rows per column.
- DataFrame.isnull().sum() = returns the number of rows per column which contains missing values

```python
missing_values_count = data.isnull().sum()
print(missing_values_count)
print(sum(missing_values_count))
```

If the goal is to remove all rows that have any missing values. Usually not the best idea as loss of information can be significant:

```python
data.dropna()  # Removes ROWS with missing values
# use .dropna(axis=1) to remove columns with missing values
```

```python
test = data.dropna(axis=1)
print(test.head())

# Illustrating the loss of data
print("Columns in original dataset: %d \n" % data.shape[1])
print("Columns with na's dropped: %d" % test.shape[1])
```

Creating a subset from the DataFrame:

```python
#  : -> All rows
#  'EPA':'Season' -> From columns EPA to Season
#  .head() -> First 5 rows from the filtered set
subset = data.loc[:, 'EPA':'Season'].head()
subset
```

In order to replace the useless NaN values, we can use the `fillna()` function to fill the missing values. We have the option to specify what NaN is replaced with:

```python
subset.fillna(0)  # This returns a new DataFrame. Doesn't modify the original one.
```

We can also replace the missing values with the value the comes directly after in the same column. This helps retain the logical progression of the data:

```python
subset.bfill(axis=0).fillna(0)  # What that remains is replaced with a 0
```

In order to find the percentage of missing datapoints in a Dataframe:

```python
missing_count = sum(data.isnull().sum())
total = data.shape[0] * data.shape[1]
percent_missing = missing_count * 100 / total
percent_missing
```
