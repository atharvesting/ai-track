```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fpath = 'iris.csv'
df = pd.read_csv(fpath)
df.head()
```

```python
sns.histplot(df['petal_length'])
```

A density plot is like a smoothed histogram. The Kernel Density Estimate (KDE) plot is an example of a density plot.

```python
sns.kdeplot(data=df.petal_length, fill=True)
```

In addition to having a single column when creating a KDE plot, we can also create a 2D KDE plot.
This can help illustrate the relationship between two parameters.

```python
sns.jointplot(x=df.petal_length, y=df.sepal_width, kind="kde")
```

```python
sns.histplot(data=df, x='petal_length', hue='class')
```

```python
sns.kdeplot(data=df, x="petal_length", hue="class", fill=True)
```
